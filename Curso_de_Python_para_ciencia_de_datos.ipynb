## Indexaci√≥n y slicing

- Indexaci√≥n: Forma de acceder a los elementos de un array por medio de su √≠ndice.

```python
array = np.random.randint(5, 30, 20)
print(array)

print("Elemento en la posici√≥n 5:", array[4])
print("Elemento en la primer posici√≥n:", array[0])
print("Elemento en la √∫ltima posici√≥n", array[-1])
print("Elemento en la ante √∫ltima posici√≥n", array[-2])
```

- Indexaci√≥n booleana: Forma de obtener datos a partir de una condici√≥n.

üì¢ Todos los elementos que cumplan con la condici√≥n ser√°n ‚Äúdevueltos‚Äù.

```python
bool_index = array > 15
print("Array booleano:", bool_index)
print("Elementos mayortes a 15:", array[bool_index])
print("Elementos entre 15 y 20:", array[(array >= 15 ) & (array <= 20)])
```

- Indexaci√≥n por listas: Permite obtener multiples elementos de un array con una lista de indices.

üì¢ Al mandarle indices ‚Äúdesordenados‚Äù el array resultante obtiene los elementos en el orden en que se pas√≥ el √≠ndice centro de la lista.

```python
index_list = [3,8,5,0]
print("Elementos en las posiciones 3,8,5 y 0:", array[index_list])
print("Elementos en las posiciones -1, 2,3 y 5:", array[[-1,2,3, 5]])
```

```python
# Indexaci√≥n de arrays multidimensionales

# Para acceder a un elemento de un array bidimensional, le indicamos 2 indices. -> `[indice_fila, indice_columna]`
matrix = np.random.randint(5, 30, (5,5))

print(matrix)
print("Elemento de la fila 2, columna 2:", matrix[2,2]) # Elemento especifio
print("Elementos de las fila 1 y 2, columna 2:", matrix[[1,2],2]) # Indexaci√≥n por lista (1)
print("Elementos en la fila 1 , columna 2, y fila 3, columna 2:", matrix[[1,2],[3,2]]) # Indexaci√≥n por lista (2)
print("Elementos mayores a 18:", matrix[matrix > 18]) # Indexaci√≥n booleana
```

- Slicing: Selecci√≥n de sub-arrays, donde le indicamos el √≠ndice inicial e √≠ndice final del conjunto original.

üì¢ Tambi√©n se se pueden agregar ‚Äúsaltos‚Äù.

```python
print("*"*5, "Array", "*"*5)
print("Elementos desde el incio hasta la posici√≥n 3:", array[:3])
print("Elementos desde la posici√≥n 5 hasta el final:", array[5:])
print("Elementos desde la posici√≥n 5 hasta la posici√≥n 10:", array[5:10])
print("Elementos desde la posici√≥n hasta la penultima posici√≥n dando 2 saltos:", array[3:-2:2])

print("*"*5, "Matriz", "*"*5)
print("Elementos a partir de la fila uno, y columnas a partir de la fila 2:\n", matrix[1:, 2:])
print("Elementos hasta la fila 3, y hasta la columna 2:\n", matrix[:3, :2])
print("Elementos a partir de la fila 1 dando 2 saltos, hasta la columna 1 dando 2 saltos\n",matrix[1::2, :1:2])
```

- Como diferenciar entre una vista y una copia

```python
original = np.array([1, 2, 3, 4])  # Crear un array original

vista = original[1:3]  # Crear una vista
copia = original.copy()  # Crear una copia

print(vista.base)  # Esto imprimir√° el array original: [1, 2, 3, 4] porque el atributo ".base" del array vista apuntar√° al array original, cuando es una vista
print(copia.base)  # Esto imprimir√° "None" porque el atributo ".base" del array vista NO apuntar√° al array original, por tratarse de una copia (otro lugar en memoria)
```

**Ejercicio**: An√°lisis de Ventas Mensuales con NumPy.

El objetivo de este ejercicio es trabajar con arrays de NumPy para analizar y manipular datos de ventas de tres productos a lo largo de un a√±o. A trav√©s de diversas operaciones, explorar√°s c√≥mo usar NumPy para obtener estad√≠sticas, realizar manipulaciones avanzadas y aplicar t√©cnicas de indexaci√≥n para extraer informaci√≥n clave.

```python
# Datos
months = np.array([
    'Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo', 'Junio',
    'Julio', 'Agosto', 'Septiembre', 'Octubre', 'Noviembre', 'Diciembre'
])

product_A = np.array([150, 200, 250, 300, 220, 210, 180, 190, 230, 240, 280, 300])
product_B = np.array([180, 210, 230, 250, 270, 260, 240, 250, 270, 290, 310, 330])
product_C = np.array([200, 220, 240, 260, 280, 300, 320, 340, 360, 380, 400, 420])
```

- Paso 2: Estad√≠sticas B√°sicas
  - Calcula la media y la suma de ventas para los productos A, B y C usando las funciones de NumPy.
  - Imprime estos valores en el formato siguiente:
  - Media de ventas Producto A: <valor>
  - Suma de ventas Producto A: <valor>
  - Repite para los productos B y C.

```python
product_A_mean = np.mean(product_A) # media de producto A
product_B_mean = np.mean(product_B) # media de producto A
product_C_mean = np.mean(product_C) # media de producto A

product_A_sum = np.sum(product_A) # suma de producto A
product_B_sum = np.sum(product_B) # suma de producto B
product_C_sum = np.sum(product_C) # suma de producto C

print("Media de ventas Producto A: " , product_A_mean)
print("Suma de ventas Producto A: ", product_A_sum)

print("Media de ventas Producto B: " , product_B_mean)
print("Suma de ventas Producto B: ", product_B_sum)

print("Media de ventas Producto C: " , product_C_mean)
print("Suma de ventas Producto C: ", product_C_sum)
```

- Paso 3: Manipulaci√≥n y An√°lisis de Datos
  - Total de ventas por mes: Suma las ventas de los tres productos para cada mes.
  - Promedio de ventas por producto: Calcula el promedio de ventas por producto.
  - Mes con mayor y menor ventas: Identifica qu√© mes tuvo el total de ventas m√°s alto y cu√°l el m√°s bajo usando las funciones np.argmax y np.argmin.

```python
# total de ventas por mes
total_ventas_mes = product_A + product_B + product_C
print(total_ventas_mes)
```

```python
# Promedio de ventas por producto
promedio_ventas_producto_A = np.mean(product_A)
promedio_ventas_producto_B = np.mean(product_B)
promedio_ventas_producto_C = np.mean(product_C)

print("Promedio de ventas Producto A: ", promedio_ventas_producto_A)
print("Promedio de ventas Producto B: ", promedio_ventas_producto_B)
print("Promedio de ventas Producto C: ", promedio_ventas_producto_C)
```

```python
# Mes con mayor y menor ventas
mes_mayor_ventas = months[np.argmax(total_ventas_mes)]
mes_menor_ventas = months[np.argmin(total_ventas_mes)]

print("Mes con mayor ventas: ", mes_mayor_ventas)
print("Mes con menor ventas: ", mes_menor_ventas)
```

- Paso 4: Operaciones Avanzadas con NumPy
  - Crea una matriz 2D con las ventas de los tres productos y transforma su forma (reshape) a un array tridimensional con dimensiones (3, 4, 3).
  - Transpone la matriz de ventas para que las filas se conviertan en columnas.
  - Invertir arrays: Invierte las ventas de cada producto en orden de meses.
  - Aplanar la matriz: Convierte la matriz de ventas a un array unidimensional.

```python
matriz_2D = np.array([product_A, product_B, product_C]) # matriz 2D
print(matriz_2D)

matriz_3D = matriz_2D.reshape(3, 4, 3) # matriz con reshape
print(matriz_3D)
```

```python
matriz_transpuesta = matriz_2D.T # matriz de ventas transpuesta
print(matriz_transpuesta)
```

```python
product_A_reversed = product_A[::-1]
product_B_reversed = product_B[::-1]
product_C_reversed = product_C[::-1]

print("Matriz de ventas de productos A en reversa: ", product_A_reversed)
print("Matriz de ventas de productos B en reversa: ", product_B_reversed)
print("Matriz de ventas de productos C en reversa: ", product_C_reversed)
```

```python
# aplanar la matriz 2d
flatten_matrix = matriz_2D.flatten()

print(flatten_matrix)
```

- Paso 5: An√°lisis de Elementos √önicos
  - Utiliza np.unique para encontrar los elementos √∫nicos en los datos de ventas y cuenta cu√°ntas veces aparece cada uno.

```python
unique_elements, counts = np.unique(flatten_matrix, return_counts=True)
print(unique_elements)
print(counts)
```

- Paso 6: Indexaci√≥n y Slicing
  - Ventas del primer trimestre: Extrae las ventas de los tres primeros meses del a√±o.
  - Indexaci√≥n booleana: Selecciona los meses donde el total de ventas de los tres productos supere los 800.
  - Selecci√≥n avanzada: Usa una lista de √≠ndices para seleccionar las ventas de los meses pares (o selecciona los meses a tu elecci√≥n) y muestra esas ventas en una nueva matriz.

## Operaciones de algebra lineal


```python
A = np.array([[1,2], [3,4]])
B = np.array([[5,6], [7,8]])

print(A)
print(B)
```

```python
# inversa de una matriz
inversa_A = np.linalg.inv(A)
print(inversa_A)
```

## GroupBy

```python
# distribuci√≥n de frecuencias es el primer paso para entender c√≥mo se distribuyen los datos

# metodo para obtener la distribucion de frecuencias de una columna
country_count = df['Country'].value_counts() # por pais cuantos registros hay

print(country_count)
```

```python
# Despu√©s de entender la distribuci√≥n de frecuencias, el siguiente paso es agrupar los datos para calcular la suma total de ventas por pa√≠s

# agrupamiento por pais que de cada uno se extrae la cantidad de ventas y se suman; de cada pais obtner la suma de las cantidades de ventas
country_group = df.groupby('Country')['Quantity'].sum()

print(country_group)
```

Para un an√°lisis m√°s profundo, calcular estad√≠sticas como la media de precios unitarios puede ofrecer informaci√≥n valiosa:

- Agrupa nuevamente los datos por pa√≠s.
- Calcula tanto la suma como la media de los precios unitarios.

```python
# agg -> agregar informacion, en este caso la media y la suma del precio unitario de cada registro por pais
country_stats = df.groupby('Country')['UnitPrice'].agg(['mean', 'sum'])

print(country_stats)
```

A veces, un solo nivel de agrupaci√≥n no es suficiente. Si deseas entender el comportamiento de las ventas no solo por pa√≠s, sino tambi√©n por stock, sigue estos pasos:

- Agrupa los datos por Country y StockCode.
- Calcula la suma de la cantidad para cada combinaci√≥n de pa√≠s y stock.

```python
# se agrupa por pais y luego por el stock code que hay en cada pais, y sumar la cantidad de registros que se manejen
country_stock_group = df.groupby(['Country', 'StockCode'])['Quantity'].sum()

print(country_stock_group)
```

Ahora para calcular el ingreso total por pais: el c√°lculo del ingresos totales depende de multiplicar la cantidad de productos vendidos por su precio unitario.

- Crea una funci√≥n que calcule el ingreso total para un grupo.
- Aplica esta funci√≥n despu√©s de agrupar los datos por pa√≠s.

```python
def total_revenue(group):
  return (group['Quantity'] * group['UnitPrice']).sum()

revenue_by_country = df.groupby('Country').apply(total_revenue) # agrupar por pais y calcular el ingreso total de ese pais en las ventas

print(revenue_by_country)
```

```python
# top 3 mejores paises en ventas

# nlargest: Devuelve los 3 resultados de mayor valor, es la combinaci√≥n de sort_values(ascending=False) y head(3).
top_best_countries = revenue_by_country.nlargest(3)

print(top_best_countries)
```

```python
# top 3 peores paises en ventas

# nsmallest: Devuelve los 3 resultados de menor valor, es la combinaci√≥n de sort_values(ascending=True) y head(3).
top_worst_countries = revenue_by_country.nsmallest(3)

print(top_worst_countries)
```

```python
revenue_by_country_sorted = revenue_by_country.sort_values(ascending=False) # ordenar de mayor a menor

print(revenue_by_country_sorted)
```

```python
revenue_by_country_sorted = revenue_by_country.sort_values(ascending=True) # ordenar de menor a mayor

print(revenue_by_country_sorted)
```

## Fusion de df

La fusi√≥n de **DataFrames** en **Pandas** es una operaci√≥n clave para combinar datos de diferentes fuentes, similar a las uniones de tablas en SQL. Se puede realizar de diferentes maneras, dependiendo de c√≥mo deseas combinar los conjuntos de datos. Las funciones principales para fusionar son merge(), join(), y concat().

```python
# primer ejemplo con merge
df1 = pd.DataFrame({
  'key': ['A', 'B', 'C'],
  'value1': [1,2,3]
})

df2 = pd.DataFrame({
  'key': ['B', 'C', 'D'],
  'value2': [4,5,6]
})

print(df1)
print(df2)
```

```python
# devolver filas con claves que coinciden para ambos df
inner_merge = pd.merge(df1, df2, on = 'key', how = 'inner')
  # on -> sobre que queremos hacer el merge
  # how -> operacion a hacer dentro, en este caso union internaa (inner)

print(inner_merge)
```

```python
# devolver todas las filas de ambos df asi no coincidan sus keys
outer_merge = pd.merge(df1, df2, on = 'key', how = 'outer') # union externa

print(outer_merge)
```

```python
# merge hacia la izquierda
left_merge = pd.merge(df1, df2, on = 'key', how = 'left') # union izquierda

print(left_merge)
```

```python
# merge hacia la derecha
right_merge = pd.merge(df1, df2, on = 'key', how = 'right') # union derecha

print(right_merge)
```

```python
# ejemplo con concatenacion
df3 = pd.DataFrame({
  'A': ['A0', 'A1', 'A2'],
  'B': ['B0', 'B1', 'B2'],
})

df4 = pd.DataFrame({
  'A': ['A3', 'A4', 'A5'],
  'B': ['B3', 'B4', 'B5'],
})

print(df3)
print(df4)
```

```python
# concatenacion vertical
vertical_concat = pd.concat([df3, df4])

print(vertical_concat)
```

```python
# concatenacion horizontal
horizontal_concat = pd.concat([df3, df4], axis = 1)

print(horizontal_concat)
```

```python
# ejemplo con join
  # Join combina df con base a un indice o columna clave

df5 = pd.DataFrame({
    'A': ['A0', 'A1', 'A2'],
    'B': ['B0', 'B1', 'B2']
  },
  index = ['K0', 'K1', 'K2']
)

df6 = pd.DataFrame({
    'C': ['C0', 'C1', 'C2'],
    'D': ['DO', 'D1', 'D2']
  },
  index = ['K0', 'K2', 'K3']
)

print(df5)
print(df6)
```

```python
joined = df5.join(df6, how = 'inner') # join interno
  # se juntan los indices que son iguales en los dos df
  # parecido a merge pero el join va por indice (en este caso)

print(joined)
```

# Proyecto aplicado

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# cargar dataset
file_path = '/content/Online_Retail.csv'
df = pd.read_csv(file_path, encoding='ISO-8859-1')

df.head()
```

- Primero se hace el EDA y la limpieza de datos

```python
print(df.info())
```

```python
print(df.describe())
```

```python
# datos faltantes
print(df.isnull().sum())
```

```python
# datos (filas) duplicadas
print(df.duplicated().sum())
```

```python
# valores unicos del df
unique_values = {column: df[column].unique() for column in df.columns}

for column, values in unique_values.items():
  print(f"Valores unicos en la columna '{column}':")
  print(f'Numero de valores unicos: {len(values)}')
  print(f'Valores unicos: {values[:10]}')
  print('-' * 50)
```

```python
# eliminar filas duplicadas
data_cleaned = df.drop_duplicates()

# eliminar valores faltantes
data_cleaned = data_cleaned.dropna(subset=['CustomerID'])
```

```python
data_cleaned.isnull().sum()
```

```python
data_cleaned.duplicated().sum()
```

- Sigue la transformaci√≥n de los datos

```python
# crear columna nueva con el total de la venta
data_cleaned['Total_Price'] = data_cleaned['Quantity'] * data_cleaned['UnitPrice']

data_cleaned.head()
```

```python
# conversion de fecha a su formato
data_cleaned['InvoiceDate'] = pd.to_datetime(data_cleaned['InvoiceDate'])

data_cleaned.info()
```

```python
# columna fecha dividida en mas columnas
data_cleaned['Year'] = data_cleaned['InvoiceDate'].dt.year
data_cleaned['Month'] = data_cleaned['InvoiceDate'].dt.month
data_cleaned['Day'] = data_cleaned['InvoiceDate'].dt.day

data_cleaned.head()
```

```python
# ventas por a√±o
sales_by_year = data_cleaned.groupby('Year')['Total_Price'].sum()

sales_by_year
```

```python
# columna semestres
data_cleaned['Semester'] = data_cleaned['Month'].apply(lambda x: 1 if x <= 6 else 2)

data_cleaned.head()
```

```python
# ventas por semestre
sales_by_semester = data_cleaned.groupby(['Year', 'Semester'])['Total_Price'].sum()

sales_by_semester
```

```python
# Ventas por mes
sales_by_month = data_cleaned.groupby(['Year', 'Month'])['Total_Price'].sum()

sales_by_month
```

- Visualizaci√≥n de datos

```python
# total de devoluciones
total_returns = data_cleaned[data_cleaned['Quantity'] < 0].shape[0]
total_returns
```

```python
# ventas sin devoluciones
total_non_returns = data_cleaned[data_cleaned['Quantity'] >= 0].shape[0]
total_non_returns
```

```python
labels = ['Devoluciones', 'No devoluciones']
sizes = [total_returns, total_non_returns]
colors = ['lightcoral', 'lightgreen']

plt.figure(figsize=(8,8))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.axis('equal')
plt.title('Proporcion de ventas con y sin devoluciones')
plt.show()
```

```python
# distribucion de ventas por mes y a√±o
plt.figure(figsize=(12,6))
data_cleaned.groupby(['Year', 'Month'])['Total_Price'].sum().plot(kind='bar', color='skyblue')
plt.title('Distribucion de ventas por mes y a√±o')
plt.xlabel('Mes y a√±o')
plt.ylabel('Ventas')

plt.show()
```

```python
# graficar productos que han sido top 10 mas vendidos
top_10_products = data_cleaned.groupby('StockCode')['Quantity'].sum().sort_values(ascending=False).head(10)
top_10_products = top_10_products.reset_index()
top_10_products = pd.merge(top_10_products, data_cleaned[['StockCode', 'Description']].drop_duplicates(), on='StockCode', how='left')

top_10_products
```

```python
plt.figure(figsize=(12,8))
plt.barh(top_10_products['Description'], top_10_products['Quantity'], color='skyblue')
plt.xlabel('Cantidad vendida')
plt.ylabel('Producto')
plt.title('Top 10 productos mas vendidos')
plt.gca().invert_yaxis()

plt.show()
```
